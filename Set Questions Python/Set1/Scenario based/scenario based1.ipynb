```python
# 1. A system checks if a user is eligible to vote based on their age.
# Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older.

Age=int(input("Enter your age:"))
if Age>=18:
    print ("Eligible")
else:
    print ("Not Eligible")
```

```python
# 2. A program processes a list of numbers and needs to find the largest value.
#  Write logic to identify and return the largest number from a given list.

numbers = [10, 25, 3, 99, 45]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

```python
# 3. A company provides employees with a 10% bonus if their salary exceeds $50,000.
#  Write logic to determine the bonus amount based on the given salary.

salary=int(input("Salary:"))
if salary>50000:
    bonus=10/100*salary
    print("Bonus amount:",bonus)
else:
    print("No Bonus")
```

```python
# 4. A program evaluates a number to determine if it is even or odd.
#  Write logic to check whether a given number is even or odd.

Number=int(input("Enter a number:"))
if Number%2==0:
    print("The number is Even")
else:
    print("The number is Odd")
```

```python
# 5. A text-processing tool reverses a given word or sentence for formatting purposes.
#  Write logic to take a word or sentence as input and produce its reversed version.


# reverse a given WORD

word = "Intelligence"
reversed_text = ""
for char in word:
    reversed_text = char + reversed_text
print("Original word:", word)
print("Reversed word:", reversed_text)

# reverse a given SENTENCE

sentence="Machine learning is one of the elements of AI "
words = sentence.split()
reversed_words = words[::-1]
reversed_sentence = " ".join(reversed_words)
print("\nOriginal sentence:", sentence)
print("Reversed sentence:", reversed_sentence)
```

```python
# 7.Scenario: A retail store offers a 20% discount if a customerâ€™s total order exceeds $100. 
#   Write logic to calculate the final amount to be paid after applying the discount.


Total_price = int(input("Enter the total price: "))

if Total_price > 100:
    final_price = Total_price - (0.20 * Total_price)
    print("Final amount after discount:", final_price)
else:
    print("No discount. Final amount:", Total_price)
```

```python
# 8. Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
# Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.

balance = 5000        
withdraw = int(input("Enter withdrawal amount:"))      

if withdraw <= balance:
    balance = balance - withdraw
    print("Withdrawal successful")
    print("Remaining balance:", balance)
else:
    print("Insufficient balance")
```

```python
# 9. Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules.
#  Write logic to determine whether a given year is a leap year.


year = int(input("Enter a year: "))  

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")

# ---A leap year must follow specific rules.---
# A year is a leap year if it is:
#     divisible by 400, OR
#     divisible by 4 but not divisible by 100.
#     Applied this logic using modulus % operator
```

```python
# 10. Scenario: A program filters out only even numbers from a given list.
#  Write logic to extract and return only the even numbers from a list.

list=[2, 13, 51, 62, 48, 29, 15, 10, 77, 32]

even_numbers = []      

for num in list:
    if num % 2 == 0:   
        even_numbers.append(num)

print("Even numbers:", even_numbers)

```
